```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
global_child_mortality_data = pd.read_csv("global-child-mortality-timeseries.csv")
global_child_mortality_data.head()
```

```python
print("No. of rows and columns:", global_child_mortality_data.shape)
```

```python
duplicate_global_child_mortality_data_rows= global_child_mortality_data[global_child_mortality_data.duplicated()]
print("No. of duplicate rows:", duplicate_global_child_mortality_data_rows.shape)
global_child_mortality_data.count()
```

```python
print(global_child_mortality_data.isnull().sum())
```

```python
global_child_mortality_data= global_child_mortality_data.rename(columns={"Share surviving first 5 years of life (based on Gapminder and World Bank (2019))":"Percentage_surviving_in_first_5_years"})
global_child_mortality_data 
```

```python
global_child_mortality_data= global_child_mortality_data.rename(columns={"Share dying in first 5 years (based on Gapminder and World Bank (2019))":"Percentage_dying_in_first_5_years"})
global_child_mortality_data   
```

```python
fig, ax = plt.subplots(figsize= (15,10))
ax.bar(global_child_mortality_data.index, global_child_mortality_data['Percentage_surviving_in_first_5_years'], label='Percentage_surviving_in_first_5_years')
ax.bar(global_child_mortality_data.index, global_child_mortality_data['Percentage_dying_in_first_5_years'], bottom=global_child_mortality_data['Percentage_surviving_in_first_5_years'], label='Percentage_dying_in_first_5_years')
ax.set_title('Child Mortality Analysis')
ax.legend()
```

```python
Country_data = pd.read_csv("country_data.csv")
Country_data.head()
```

```python
print("No. of rows and columns:", Country_data.shape)
```

```python
duplicate_Country_data_rows= Country_data[Country_data.duplicated()]
print("No. of duplicate rows:", duplicate_Country_data_rows.shape)
Country_data.count()
```

```python
print(Country_data.isnull().sum())
```

```python
Country_data_new = Country_data.drop(['Region', 'Population', 'Area (sq. mi.)', 'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)', 'Net migration', 'Literacy (%)', 'Phones (per 1000)', 'Arable (%)', 'Crops (%)', 'Other (%)', 'Climate', 'Birthrate', 'Deathrate', 'Agriculture', 'Industry', 'Service'], axis=1)
Country_data_new
```

```python
Country_data_new.dtypes
```

```python
Country_data_new['Infant mortality (per 1000 births)'] = Country_data_new['Infant mortality (per 1000 births)'].str.replace(',', '.')
Country_data_new = Country_data_new.astype({'Infant mortality (per 1000 births)': np.float})
Country_data_new
```

```python
Country_data_new.dtypes
```

```python
Country_data_new1= Country_data_new.dropna(subset=["Infant mortality (per 1000 births)","GDP ($ per capita)"])
Country_data_new1
```

```python
#Sorting the Infant Mortality column into descending order and using cmap to highlight the contrast in Infant mortality and Gdp of different countries

Country_data_new1= Country_data_new1.sort_values(by="Infant mortality (per 1000 births)", ascending=False)
Country_data_new1.style.background_gradient(cmap='Oranges')
```

```python
#The above cmap shows the contrast between the Infant mortality and the GDP of different countries.

plt.rcParams['figure.figsize'] = [20,10]
plt.plot(Country_data_new1['GDP ($ per capita)'], Country_data_new1['Infant mortality (per 1000 births)'])
plt.title('Contrast between the Infant mortality and the GDP of different countries')
plt.xlabel('GDP ($ per capita)')
plt.ylabel('Infant mortality (per 1000 births)')
plt.show()
```

```python
Country_data_new1['Infant mortality (per 1000 births)'].corr(Country_data_new1['GDP ($ per capita)'])
```

```python
#The negative correlation shows that large values of Infant Mortality correspond to small values of the GDP, and vice versa.
```
